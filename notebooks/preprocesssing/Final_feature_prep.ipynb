```python
#AGEDATA
columns_to_drop_age = ['up_votes','down_votes', 'accent', 'gender', 'duration','text']
df_dropped3 = df.drop(columns=columns_to_drop_age)
age_df = df_dropped3.dropna().reset_index(drop=True)
```

```python
#ACCENTDATA
columns_to_drop_accent = ['up_votes','down_votes', 'age', 'gender', 'duration','text']
df_dropped4 = df.drop(columns=columns_to_drop_accent)
accent_df = df_dropped4.dropna().reset_index(drop=True)
```

```python
#Age preprocessing
```

```python
filtered_age_df = age_df[age_df['age'].isin(['twenties', 'thirties', 'fourties', 'fifties', 'sixties'])]
```

```python
# Get the counts after filtering
age_counts = filtered_age_df['age'].value_counts()
print("Filtered age counts:")
print(age_counts)

# Find the minimum count (which should be 'sixties' with 8556)
min_count = age_counts.min()  # Should be 8556

# Create a balanced dataframe by sampling from each age group
balanced_age_df = pd.DataFrame()

for age_group in ['twenties', 'thirties', 'fourties', 'fifties', 'sixties']:
    # Get all rows for this age group
    age_group_df = filtered_age_df[filtered_age_df['age'] == age_group]

    # Sample exactly min_count rows from this group (with random seed for reproducibility)
    sampled_df = age_group_df.sample(n=min_count, random_state=42, replace=False)

    # Add to the balanced dataframe
    balanced_age_df = pd.concat([balanced_age_df, sampled_df])

# Verify the counts are now equal
print("\nBalanced age counts:")
print(balanced_age_df['age'].value_counts())

# Reset the index of the final dataframe
balanced_age_df = balanced_age_df.reset_index(drop=True)
```

```python
#Accent preprocessing
```

```python
accent_df['accent'].value_counts()
```

```python
# First, filter to keep only accents up to 'ireland' in your list
selected_accents = ['us', 'england', 'indian', 'australia', 'canada', 'scotland', 'newzealand', 'african', 'ireland']
filtered_accent_df = accent_df[accent_df['accent'].isin(selected_accents)]

# Get the counts after filtering
accent_counts = filtered_accent_df['accent'].value_counts()
print("Filtered accent counts:")
print(accent_counts)

# Find the minimum count (which should be 'ireland' with 1860)
min_count = accent_counts.min()  # Should be 1860
print(f"\nMinimum count: {min_count}")

# Create a balanced dataframe by sampling from each accent group
balanced_accent_df = pd.DataFrame()

for accent_group in selected_accents:
    # Get all rows for this accent group
    accent_group_df = filtered_accent_df[filtered_accent_df['accent'] == accent_group]

    # Sample exactly min_count rows from this group (with random seed for reproducibility)
    sampled_df = accent_group_df.sample(n=min_count, random_state=42, replace=False)

    # Add to the balanced dataframe
    balanced_accent_df = pd.concat([balanced_accent_df, sampled_df])

# Verify the counts are now equal
print("\nBalanced accent counts:")
print(balanced_accent_df['accent'].value_counts())

# Reset the index of the final dataframe
balanced_accent_df = balanced_accent_df.reset_index(drop=True)
```

```python
gender_df['gender'].value_counts()
```

```python
# First, filter to keep only 'male' and 'female'
selected_genders = ['male', 'female']
filtered_gender_df = gender_df[gender_df['gender'].isin(selected_genders)]

# Get the counts after filtering
gender_counts = filtered_gender_df['gender'].value_counts()
print("Filtered gender counts:")
print(gender_counts)

# Find the minimum count (which should be 'female' with 35095)
min_count = gender_counts.min()  # Should be 35095
print(f"\nMinimum count: {min_count}")

# Create a balanced dataframe by sampling from each gender group
balanced_gender_df = pd.DataFrame()

for gender_group in selected_genders:
    # Get all rows for this gender group
    gender_group_df = filtered_gender_df[filtered_gender_df['gender'] == gender_group]

    # Sample exactly min_count rows from this group (with random seed for reproducibility)
    sampled_df = gender_group_df.sample(n=min_count, random_state=42, replace=False)

    # Add to the balanced dataframe
    balanced_gender_df = pd.concat([balanced_gender_df, sampled_df])

# Verify the counts are now equal
print("\nBalanced gender counts:")
print(balanced_gender_df['gender'].value_counts())

# Reset the index of the final dataframe
balanced_gender_df = balanced_gender_df.reset_index(drop=True)
```
